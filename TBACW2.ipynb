# Tokenization

```python
import regex as re  # Import the 'regex' library and alias it as 're'

def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

```python
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
print(testtokens)
```

# Removal of stop words

```python
import nltk  # Import the NLTK library and download the 'stopwords' dataset for English.
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english')) # Create a set of English stopwords using NLTK's corpus module
print(stopwords) # Create a set of English stopwords using NLTK's corpus module

```

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]#Just testing it
cleantokens= remove_stop(testtokens)
print(cleantokens)

```

```python
def remove_stop(tokens): # Define a function 'remove_stop' that takes a list of tokens as input.
    return [t for t in tokens if t.lower() not in stopwords] # Use a list comprehension to filter out tokens that are in lowercase and not in the 'stopwords' set.
#Just testing it
cleantokens= remove_stop(testtokens) # Just testing the 'remove_stop' function with a list of tokens called 'testtokens'.
print(cleantokens) # Print the result of removing stopwords from 'testtokens'.
```

```python
pipeline = [str.lower, tokenize, remove_stop] # Define a list 'pipeline' that contains three functions: str.lower, tokenize, and remove_stop.

def prepare(text, pipeline): # Define a function 'prepare' that processes text using the specified pipeline of functions.
    tokens = text  # Initialize 'tokens' with the input text.
    for fun in pipeline:  # Iterate through each function in the 'pipeline' and apply it sequentially to 'tokens'.
        tokens = fun(tokens)
    return tokens # Return the final processed 'tokens'

```

```python
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline) # Add a new column 'tokens' to the DataFrame 'df' by applying the 'prepare' function to the 'text' column using the specified 'pipeline' of functions.
```

```python
df['num_tokens'] = df['tokens'].map(len) # Add a new column 'num_tokens' to the DataFrame 'df' that contains the count of tokens in each row's 'tokens' column.
df.tail() # Add a new column 'num_tokens' to the DataFrame 'df' that contains the count of tokens in each row's 'tokens' column.

```

```python
from collections import Counter # Import the 'Counter' class from the 'collections' module.

tokens = tokenize("She likes my cats and my cats like my sofa.") # Tokenize the given text and store the tokens in the 'tokens' variable.

testcount= Counter(tokens) # Create a Counter object 'testcount' to count the occurrences of each token.
print(testcount) # Print the updated 'testcount', which now includes counts for both sets of tokens.
```

```python
more_tokens = tokenize("She likes dogs and cats.") # Update the 'testcount' Counter object by adding the counts of tokens from 'more_tokens'.
testcount.update(more_tokens) # Update the 'testcount' Counter object by adding the counts of tokens from 'more_tokens'.
print(testcount) # Print the 'testcount' Counter object, which now contains the frequency counts of tokens.
```

```python
counter = Counter() # Tokenize the new text "She likes dogs and cats." and store the tokens in the 'more_tokens' variable.
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

```python
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq']) # Create a DataFrame from a dictionary 'counter' with token frequencies as values.
freq_df = freq_df.query('freq >= 1') # Filter the DataFrame to include only tokens with a frequency greater than or equal to 1.
freq_df.index.name = 'token' # Set the name of the index column to 'token'.
```

```python
freq_df = freq_df.sort_values('freq', ascending=False) # Sort the DataFrame 'freq_df' based on the 'freq' column in descending order (highest frequency first).
freq_df.head(10) # Display the top 10 rows (tokens with the highest frequencies) from the sorted DataFrame.
```
